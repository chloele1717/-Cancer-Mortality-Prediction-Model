**[2.5]** Handle outliners: Usually z-score =3 is considered as a cut-off value to set the limit. Therefore, any z-score greater than +3 or less than -3 is considered as outlier which is pretty much similar to standard deviation method

For training set

```
# calculate the z-scores for each numeric column
z_scores = np.abs((df_cleaned - df_cleaned.mean()) / df_cleaned.std())

# create a boolean mask for values that are more than 3 standard deviations from the mean
mask = (z_scores > 3).any(axis=1)

# remove the rows that contain outliers
df_cleaned = df_cleaned[~mask]
```

For testing shape

```
# calculate the z-scores for each numeric column
z_scores_test = np.abs((df_test_cleaned - df_test_cleaned.mean()) / df_test_cleaned.std())

# create a boolean mask for values that are more than 3 standard deviations from the mean
mask_test = (z_scores_test > 3).any(axis=1)

# remove the rows that contain outliers
df_test_cleaned = df_test_cleaned[~mask_test]
```
